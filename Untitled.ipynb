```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
!pip install wordcloud
```

```python
df = pd.read_csv('Amazon_bestsellers_items_2025.csv')
```

```python
df['product_price'] = (
    df['product_price']
    .astype(str)                                # ensure string
    .str.replace(r'[^0-9.]', '', regex=True)    # remove $, commas, etc.
    .replace('', np.nan)                        # replace empty string with NaN
    .astype(float)                              # convert to float safely
)

print(df['product_price'].head(10))
print(df['product_price'].describe())
```

```python
top_n = 10  
top_products = df.sort_values('rank').head(top_n)

columns_to_show = ['rank', 'product_title', 'product_price', 'product_star_rating', 'product_num_ratings']
print(top_products[columns_to_show])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['product_price'], bins=50, kde=True)
plt.title("Overall Price Distribution")
plt.show()

# Price distribution by country
plt.figure(figsize=(10,6))
sns.boxplot(x='country', y='product_price', data=df)
plt.title("Price Distribution by Country")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average star rating by country
print(df.groupby("country")["product_star_rating"].mean().sort_values(ascending=False))

# Top rated products
print(df.sort_values("product_star_rating", ascending=False)[["product_title", "product_star_rating"]].head(10))

# Products with most reviews
print(df.sort_values("product_num_ratings", ascending=False)[["product_title", "product_num_ratings"]].head(10))

# Relationship between ratings and number of reviews
plt.figure(figsize=(8,6))
sns.scatterplot(x="product_num_ratings", y="product_star_rating", data=df, alpha=0.6)
plt.title("Relationship: Reviews vs Rating")
plt.show()
```

 RANKING TRENDS & CHANGES

```python
if "date" in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    product = df['product_title'].iloc[0]
    product_rank = df[df['product_title']==product]

    plt.figure(figsize=(10,6))
    sns.lineplot(x="date", y="rank", data=product_rank, marker="o")
    plt.title(f"Rank Trend for {product}")
    plt.gca().invert_yaxis()
    plt.show()
```

 COUNTRY-WISE MARKET INSIGHTS

```python
print("\nTop products by country:")
top_products_country = df.groupby("country")["product_title"].value_counts().groupby(level=0).head(5)
print(top_products_country)

print("\nUnique products per country:")
unique_country_products = df.groupby("country")["product_title"].unique()
print(unique_country_products)

print("\nPrice comparison of same products across countries:")
same_product_prices = df.groupby(["product_title", "country"])["product_price"].mean().unstack()
print(same_product_prices.head(10))
```

Category Analysis

```python
def categorize(title):
    title = title.lower()
    if "antivirus" in title:
        return "Antivirus"
    elif "office" in title or "word" in title or "excel" in title:
        return "Office Suite"
    elif "vpn" in title:
        return "VPN"
    elif "video" in title or "editor" in title:
        return "Video Editing"
    else:
        return "Other"

df["category"] = df["product_title"].apply(categorize)

plt.figure(figsize=(10,6))
sns.countplot(x="category", data=df, order=df["category"].value_counts().index)
plt.title("Category Distribution")
plt.xticks(rotation=45)
plt.show()

print("\nCategory share by country:")
category_counts = df.groupby("country")["category"].value_counts(normalize=True).unstack().fillna(0)
print(category_counts)
```

CORRELATION ANALYSIS

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["product_price","product_star_rating","product_num_ratings","rank"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

sns.scatterplot(x="product_price", y="product_star_rating", data=df)
plt.title("Price vs Rating")
plt.show()

sns.scatterplot(x="product_price", y="product_num_ratings", data=df)
plt.title("Price vs Number of Ratings")
plt.show()

sns.scatterplot(x="product_star_rating", y="rank", data=df)
plt.title("Star Rating vs Rank")
plt.show()
```

 TIME-BASED SALES INSIGHTS

```python
if "date" in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.to_period('M')

    sales_trend = df.groupby("month")["product_title"].count()
    sales_trend.plot(kind="line", figsize=(10,6), marker="o")
    plt.title("Monthly Sales Trend")
    plt.show()
```
